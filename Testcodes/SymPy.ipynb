# SymPy Usage!
___

```python
# Import everything about SymPy
from sympy import *
import numpy as np
# LaTeX formatted
init_printing()
```

```python
# Define your symbols
x, y = symbols("x y")
a, b, c, d = symbols("a b c d")
```

```python
# Define your expressions
expr1 = x**2 + 2*x*y + y**2
expr2 = (x + y)**2
```

## Manipulating expressions: Expand, factor, simplify!

```python
# Expanding, factoring, and simplifying
expanded = expand(expr2)
factored = factor(expanded)
simplified = simplify((expr1/expr2))
```

```python
# Displaying them 
display(expanded, factored, simplified)
# or 
expanded, factored, simplified
```

## Substituting values! 

```python
# Substitute values
val = expr1.subs({x:2, y:3})
val
```

```python
# Substitute expressions
new_expr1 = expr1.subs(x, y + 1)
new_expr1
```

## Solving equations symbolically!

```python
# Define the equation
eq = Eq(x**2 - 4, 0)

# Solve for x
sols = solve(eq, x)
sols
```

```python
eq = Eq(x**2 + 4, 0)
sols = solve(eq, x)
sols
```

```python
# A set of equations
eq1 = Eq(x + y, 5)
eq2 = Eq(x - y, 1)

# Solve for x and y
sol = solve((eq1, eq2), (x, y))
sol
```

## Calculus!

### Differentiation

```python
# Define expression
expr3 = x**3 + 2*x**2 

# Take 1st derivative
der3_1 = diff(expr3, x)
display(der3_1)

# Take 2nd derivative
der3_2 = diff(expr3, x, 2)
display(der3_2)
```

### Integration

```python
# Indefinite integral
indefi = integrate(sin(x), x)
display(indefi)

# Definite integral
defi = integrate(sin(x), (x, 0, np.pi))
display(defi)
```

### Limits

```python
# Limit
lim = limit((1 + 1/x)**x, x, float('inf'))
display(lim)
```

## Matrices, Linear Algebra!

```python
# Setting up a matrix
M = Matrix([[a, b], [c, d]])
M
```

### Operations

```python
# Transpose
M.T
```

```python
# Determinant
M.det()
```

```python
# Inverse
M.inv()
```

```python
# Eigenvalues and Eigenvectors
eigenval = M.eigenvals()
eigenvec = M.eigenvects()

display(eigenval)
display(eigenvec)
```

```python
# Example
M1 = Matrix([[2, 0], [0, 3]])
display(M1)
eigenval_M1 = M1.eigenvals()
display(eigenval_M1)
eigenvec_M1 = M1.eigenvects()
display(eigenvec_M1)
# Output tells us {Eigenvalue1 : Alg.Mult.1, Eigenvalue2 : Alg.Mul.2}
# Also {Eigenvalue1, Alg.Mult.1, Eigenvector(s), Eigenvalue2, Alg.Mul.2, Eigenvector(s)}
```

```python
# Multiplication
V = M * M1
display(V)
```
